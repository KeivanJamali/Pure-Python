```python
data = pd.read_csv("data.csv")
real_data = pd.read_csv("data_real.csv").iloc[:, 1]
real_data.index = list(range(1, len(real_data)+1))
model = Forcasting_Formulas(data=data)
print(model.data.head(10))
model.plotting(data=data, data_name="Data")
```

### Simple Moving Average - 3 years

```python
name = "Simple Moving Average - 3 years"
model = Forcasting_Formulas(data=data)
small_data, total_data = model.simple_moving_average(number_of_data=3, number_of_prediction=5)
model.plotting(data=total_data, data_name=name)
print(small_data)
score_SMA_3 = model.score(method=["MAE", "MSE", "MAD", "R2"], true_val=real_data, predict_val=small_data)
print(score_SMA_3)
```
