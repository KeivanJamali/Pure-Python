```python
from IPython.display import Markdown, display
from pdf2image import convert_from_path
import os
```

```python
def pdf_to_images(pdf_list, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # convert pdf to images.
    images_group = []
    for pdf in pdf_list:
        result_images = convert_from_path(pdf)
        images_group.append(result_images)

    images_paths = []
    # save images
    for i, images in enumerate(images_group):
        for j, image in enumerate(images):
            image_path = os.path.join(output_folder, f"page_{i+1}_{j+1}.jpg")
            image.save(image_path, "JPEG")
            images_paths.append(image_path)

    return images_paths
```

```python
from colorama import Fore

extracted_documents = []

for image_path in image_paths[:10]:
    print(Fore. BLUE + f"processing image" + Fore.RESET)
    with open(image_path, "rb") as image_file:
        image_data = base64.b64encode(image_file.read()).decode("utf-8")

    response = client_groq.chat.completions.create(
    model=model_groq,
    messages=[
        {"role": "user", "content": [{"type": "text",
                                      "text": "Rewrite all the information of the page, starts with: 'content: '"},
                                     {"type": "image_url",
                                      "image_url": {"url": f"data:image/jpeg;base64,{image_data}",
                                                    "detail": "high"}}]
        }
    ],
    temperature=0
    )
    
    extracted_documents.append({"image_path": image_path,
                                "document": response.choices[0].message.content})
        
    print(Fore.RED + f"Extracted information for {image_path[-10:]}\n" + Fore.GREEN + f"Message: {response.choices[0].message.content}\n" + Fore.RESET)

```

```python
# filtered the data.
filtered_document = []
for doc in extracted_documents:
    if any(word in doc["document"].lower() for word in ["instructions", "Data", "understanding", "general", "im"]):
        filtered_document.append(doc)
    else:
        print(f"skipping data:{doc["image_path"]}")
```
