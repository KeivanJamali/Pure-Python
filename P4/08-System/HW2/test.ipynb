```python
import toolkit
import sympy as sp
```

```python
def objective_function(x):
    r1 = 2 - x/3
    r2 = (x-3)**2
    r3 = x/5
    return max([r1, r2, r3])

model = toolkit.Golden_Section_Method(objective_function, domain=[0, 8])

model.fit(error=0.15)
```

```python
model.history
```

```python
def objective_function(x):
    r1 = 3 * x**2 - 4*x + 1
    return r1

model = toolkit.Bisection_Method(objective_function, domain=[0, 2])

model.fit(error=0.1)
```

```python
model.history
```

```python
x1, x2 = sp.symbols("x_1, x_2")
objective_func = (x1-2)**2 + 4*(x2-3)**2

model = toolkit.Steepest_Descent_Method(objective_function=objective_func, parameters=[x1, x2])
```

```python
model.fit(x0=[0, 0], error=0.1)
```

```python
model.plot_contour(xlim=[-1, 5], ylim=[0, 5], resolution=50)
```
