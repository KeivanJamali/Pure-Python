```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Correlation Coefficient

```python
id = np.linspace(1, 100, 100)
def make_y1(x):
    y = np.tan(np.pi/3) * x + 1
    return y

def make_y2(x):
    y = np.tan(np.pi/3) * x + 25
    return y

y1 = make_y1(id)
y2 = make_y2(id)
```

```python
plt.plot(id, y1)
plt.plot(id, y2)
```

```python
correlation_matrix = np.corrcoef(y1, y2)
```

```python
def covariance_v0(y1, y2):
    E_y1 = sum(y1) / len(y1)
    E_y2 = sum(y2) / len(y2)
    E_y1y2 = sum(y1[i]*y2[i] for i in range(len(y1))) / len(y1)

    cov_y1_y2 = E_y1y2 - E_y1 * E_y2
    return cov_y1_y2

def covariance_v1(y1, y2):
    y1_mean = sum(y1) / len(y1)
    y2_mean = sum(y2) / len(y2)

    cov_y1_y2_sum = sum((y1[i] - y1_mean)*(y2[i] - y2_mean) for i in range(len(y1)))
    cov_y1_y2 = cov_y1_y2_sum / len(y1)
    return cov_y1_y2

def correlation_v0(y1, y2):
    cov_y1_y2 = covariance_v0(y1, y2)
    var_y1 = covariance_v0(y1, y1)
    var_y2 = covariance_v0(y2, y2)
    cor_y1_y2 = cov_y1_y2 / (np.sqrt(var_y1*var_y2))
    return cor_y1_y2
```
