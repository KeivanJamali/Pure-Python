```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
p, q = sp.symbols("p, q")
q1 = 5.876 * (10 ** 8) * p **(-2.75)
q2 = 46875 - 687.5 * p
```

```python
x_values = np.linspace(20, 100, 1000)
lambdify1 = sp.lambdify(p, q1, "numpy")
lambdify2 = sp.lambdify(p, q2, "numpy")

y_values1 = lambdify1(x_values)
y_values2 = lambdify2(x_values)
```

```python
plt.scatter(x_values, y_values1, marker=".", label="q=a-B*p")
plt.scatter(x_values, y_values2, marker=".", label="q=ap**(-B)")
plt.title("Assume point (50, 12500)")
plt.legend()
plt.savefig("Kraft VS Linear")
```
