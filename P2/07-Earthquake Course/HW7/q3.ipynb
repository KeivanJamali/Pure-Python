```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def b3_func(t, data):
    s0, s, t0, ts = data
    if 0<=t<t0:
        return 1+ s *(t/t0)
    elif t0<=t<ts:
        return s+1
    elif t>=ts:
        return (s+1)*((ts/t)**(2/3))
    else:
        raise TypeError("[ERROR] Change t...")

def b4_func(t, data):
    s0, s, t0, ts = data
    if 0<=t<t0:
        return s0 + (s-s0+1)*(t/t0)
    elif t0<=t<ts:
        return s+1
    elif t>=ts:
        return (s+1)*(ts/t)
    else:
        raise TypeError("[ERROR] Change t...")
```

```python
data = [[1, 1.5, 0.1, 0.4], [1, 1.5, 0.1, 0.4], 
        [1, 1.5, 0.1, 0.5], [1, 1.5, 0.1, 0.5], 
        [1.1, 1.75, 0.15, 0.7], [1.1, 1.75, 0.15, 0.7], 
        [1.3, 2.25, 0.15, 1], [1.1, 1.75, 0.15, 1]]
names = ["LowMedium_I", "High_I", "LowMedium_II", "High_II", "LowMedium_III", "High_III", "LowMedium_IV", "High_IV"]
c_name = "High_III"
x = np.arange(0, 4.5, 0.01)
result_df = pd.DataFrame({"Period(sec)": x}, index=x)
result_df
```

```python
for i in range(len(names)):
    result_df[names[i]+"-edition3"] = result_df["Period(sec)"].apply(lambda x: b3_func(x, data[i]))
    result_df[names[i]+"-edition4"] = result_df["Period(sec)"].apply(lambda x: b4_func(x, data[i]))
```

```python
result_df[[c_name+"-edition4"]].plot();
print(c_name+"-edition4")
```
