```python
from modular import dataloader, models, gravity
import numpy as np
import pandas as pd
```

```python
city_path = r"/mnt/Data1/Python_Projects/Pure-Python/P5/01-Thesis/data/Anaheim"
data = dataloader.Custom_DataLoader(city_name="Anaheim", city_path=city_path)
data.fit()
```

```python
data.data
```

```python
import numpy as np

def exponential_decay(d, beta):
    return np.where(d > 0, np.exp(-beta * d), 0)

model = gravity.GravityModel(decay_function=exponential_decay, param_names=["beta"])

model.fit(data.data, decay_params={"beta": 0.1})

# Or calibrate the parameter
params = model.calibrate(data.data, initial_params={"beta": 0.1}, bounds=[(0.01, 200)])
print("Best beta:", params["beta"])
```

```python
import torch.nn as nn
from torch import optim
from tqdm.auto import tqdm

model_gravity = models.GravityNet()
optimizer = optim.Adam(model_gravity.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

# Training
for epoch in tqdm(range(100)):
    for x, y in data.dataloader:
        optimizer.zero_grad()
        y_hat = model_gravity(x[:, 0], x[:, 1], x[:, 2])
        loss = loss_fn(y_hat, y)
        loss.backward()
        optimizer.step()
    
    if epoch % 1 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}, Beta = {model_gravity.beta.item():.4f}, k = {model_gravity.k.item():.4f}")
```
