```python
import matplotlib.pyplot as plt
import numpy as np

# Time intervals (hours)
time = np.arange(0, 16, 0.1)  # from 6:00 AM to 10:00 PM (16 hours)

# Queue length over time
queue_length = []

# Calculate queue length for each time point
for t in time:
    if t <= 8:  # From 6:00 AM to 2:00 PM
        queue_length.append(4 * t)
    elif t <= 12:  # From 2:00 PM to 6:00 PM
        queue_length.append(32 - 8 * (t - 8))
    else:  # From 6:00 PM to 10:00 PM
        queue_length.append(0)

# Plotting the graph
plt.figure(figsize=(12, 6))
plt.plot(time, queue_length, label='Queue Length', color='blue')
plt.axvline(x=8, color='red', linestyle='--', label='2:00 PM')
plt.axvline(x=12, color='green', linestyle='--', label='6:00 PM')
plt.xlabel('Time (hours after 6:00 AM)')
plt.ylabel('Number of Trucks in Queue')
plt.title('Queue Length Over Time')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig("Q3.svg", format="svg")
```
