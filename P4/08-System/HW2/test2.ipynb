```python
from Convex_Combination import ConvexCombinationOptimizer
import sympy as sp

x1, x2 = sp.symbols("x1 x2")
objective_function = 4*(x1 - 10)**2 + (x2 - 4)**2
constraints = [[10, 1, -1], [3, 0.2, -1]]
equality = ["leq", "geq"]
parameters = [("x1", "+"), ("x2", "n")]

model = ConvexCombinationOptimizer(objective_function=objective_function,
                                   params=[x1, x2],
                                   constraint_set=constraints,
                                   equality=equality,
                                   parameters=parameters)

model.fit(x0=[0, -3], lr=None, error=0.01)
model.plot_contour(xlim=[0, 10], ylim=[-5, 0])
```
