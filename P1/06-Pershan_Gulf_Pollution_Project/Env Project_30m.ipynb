# Entering Data and analysis

```python
m = 1000  # kg
d = 68  # m
dx = 4.5 * 10 ** -5  # m^2 /s
dy = 6.5 * 10 ** -5  # m^2 /s
distance = 100  # m
time_start = 1  # hour
time_end = 5  # hour
q1 = te.Transport(m, dx, dy)
color = []
for i in colors_kit.color4:
    for j in i:
        for k in range(7):
            color.append(j)
print(len(color))
```

```python
distance = 30//2  # m
time_scale = 1  # hour
time_scale = int(1 / time_scale)
time_nodes = []
iteration = -1
for t in range(time_start * time_scale, (time_end + 1) * time_scale):
    t /= time_scale
    t *= 3600
    iteration += 1
    time_nodes.append([])
    for x in range(-distance * 10, distance * 10):
        x /= 10
        for y in range(-distance * 10, distance * 10):
            y /= 10
            c = q1.diffusion(2, t, x, y, L=d)
            time_nodes[iteration].append(te.Point(x, y, c=c))
```

# from 1 to 5 hour plot_curve in scenario 1

```python
for i in range(len(time_nodes)):
    t = time_start + (time_end - time_start) * (i / (len(time_nodes) - 1))
    te.plot_curve(time_nodes[i], t, scenario=1, show=False, save=True, scale=distance)
```

# from 1 to 5 hour plot_picture in scenario 1

```python
for i in range(len(time_nodes)):
    t = time_start + (time_end - time_start) * (i / (len(time_nodes) - 1))
    te.plot_picture(t, time_nodes[i], scale=distance, color=color, error=0.001, scenario=1, save=True, show=True, detail=False)
```
