```python
%load_ext autoreload
%autoreload 2
from TR1 import Plot
import sympy as sp
```

## Simple Usage

```python
p = Plot(use_seaborn=True)
p.add_line(sp.sympify('x**2 - 2'), label='Parabola')
p.add_point(10, 3, label="A")
p.add_segment(2, 3, -1, 1, label="Segment A-B")
p.add_inequality(sp.sympify('12*x + 2*y <= 12'), label='Inequality')
p.show()
```

## Question 1

```python
p = Plot(use_seaborn=True, title="Z=2x + y")
obj_func = sp.sympify('2*x + y')
constraints = [sp.sympify('12*x + 3*y <= 6'), sp.sympify('-3*x + y <= 7'), sp.sympify('y <= 10'), sp.sympify('x >= 0'), sp.sympify('y >= 0')]
p.plot_optimization(obj_func, constraints, minimize_obj=False)
p.set_plot_limits(x_min=-5, x_max=5, y_min=-5, y_max=11)
p.show()
p.save("Figs/Q1-a.svg")
```
