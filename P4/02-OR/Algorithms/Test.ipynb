# Shortest Path Problem

### Book Example

```python
import Shortest_Path_Problem as spp
# For nodes we have the name and the position.
nodes = {"A": (0, 0),
         "B": (1, 1),
         "C": (1, -1),
         "D": (2, 1.5),
         "E": (2, 0),
         "F": (2, -1.5),
         "G": (3, 1),
         "H": (3, -1),
         "I": (4, 0)}

# for edges we need only the edge
edges = {("A", "B") : [7],
         ("A", "C") : [9],
         ("B", "C") : [3],
         ("B", "D") : [10],
         ("B", "E") : [14],
         ("C", "E") : [10],
         ("C", "F") : [6],
         ("D", "G") : [8],
         ("E", "G") : [9],
         ("E", "I") : [13],
         ("E", "H") : [4],
         ("E", "F") : [2],
         ("F", "H") : [7],
         ("G", "I") : [10],
         ("H", "I") : [6]}

problem = spp.Defind_Graph()
problem.set_setting(nodes=nodes, edges=edges)
```

```python
problem.plot_graph()
```

```python
model = spp.Dijkestra(problem=problem)
model.fit(start_node="G", end_node="I")
# model.extract_solution(node="I")
```

### HomeWork Example

```python
import Shortest_Path_Problem as spp
# For nodes we have the name and the position.
nodes = {"1": (-1, 0),
         "2": (1, 0),
         "3": (3, 0),
         "4": (2, -1),
         "5": (1, -1),
         "6": (0, -1)}

# for edges we need only the edge
edges = {("1", "2") : [10],
         ("1", "5") : [4],
         ("1", "6") : [15],
         ("2", "3") : [2],
         ("2", "4") : [3],
         ("2", "6") : [5],
         ("3", "4") : [1],
         ("3", "5") : [3],
         ("4", "5") : [6]}

problem = spp.Defind_Graph()
problem.set_setting(nodes=nodes, edges=edges)
problem.plot_graph()
```

# Minimum Spanning Tree Problem

### Book Example 1

```python
import Minimum_Spanning_Tree_Problem as mstp
# For nodes we have the name and the position.
nodes = {"A": (1, 5),
         "B": (3, 4),
         "C": (5, 5),
         "D": (2, 3),
         "E": (4, 3),
         "F": (3, 2),
         "G": (6, 1)}

# for edges we need only the edge
edges = {("A", "B") : [7],
         ("A", "D") : [5],
         ("B", "C") : [8],
         ("B", "D") : [9],
         ("B", "E") : [7],
         ("C", "E") : [5],
         ("D", "E") : [15],
         ("D", "F") : [6],
         ("E", "G") : [9],
         ("E", "F") : [8],
         ("F", "G") : [11]}

problem = mstp.Defind_Graph()
problem.set_setting(nodes=nodes, edges=edges)
problem.plot_graph()
```

```python
model = mstp.Kruskal(problem=problem)
model.fit()
```
