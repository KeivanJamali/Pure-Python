# Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import NormalDist
```

```python
x = np.linspace(-5, 5, 100)
normal = NormalDist(mu=0, sigma=1)
y_cdf = []
y_pdf = []
for i in x:
    y_cdf.append(normal.cdf(i))
    y_pdf.append(normal.pdf(i))
plt.scatter(x, y_cdf, label="Cumulative Distribution Function", marker=".")
plt.scatter(x, y_pdf, label="Probability Density Function($\phi$)", marker=".")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$p$")
```

```python
x = np.linspace(-2, 2, 100)
normal = NormalDist(mu=0, sigma=1)
y_cdf = []
y_new = []
for i in x:
    y_cdf.append(normal.cdf(i))
    y_new.append(1 - normal.cdf(i))
plt.scatter(x, y_cdf, label="CDF(1 - $\Phi$)", marker=".")
plt.scatter(x, y_new, label="$\Phi$", marker=".")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$p$")
plt.show()
```

```python
normal = NormalDist(mu=0, sigma=1)
decimal = 8

Q = 2472.38
h = 0.4
p = 1
landa = 12000
mu = 400
sigma = 50
k = 100

print(f"Start with Q = {Q}")

p_for_r = (Q*h)/(p*landa)
R_standard = normal.inv_cdf(1-p_for_r)
R = round(R_standard * sigma + mu, decimal)
print(f"R value is calculated to be {R}")
#-----------------
val = (R - mu)/sigma
y1 = round(normal.pdf(val), decimal)
y2 = round((1 - normal.cdf(val)), decimal)
final_output = round(sigma * (y1) - (R - mu) * (y2), decimal)

print(f"PDF for point: {val} = ({R} - {mu})/{sigma} is : {y1}")
print(f"(1 - CDF) for point: {val} = ({R} - {mu})/{sigma} is : {y2}")
print(f"Final output for the E[(D-R)^+] is : {final_output}")

Q = (2 * landa * (k+ p * final_output)) / h
Q = round(np.sqrt(Q), decimal)

print("\n\n",6*"*", f"Final Q is : {Q}", 6*"*")
```
