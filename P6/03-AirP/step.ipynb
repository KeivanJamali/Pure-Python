```python
import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Add edges with distances
edges = [
    ('A', 'B', 2000),
    ('A', 'C', 4200),
    ('A', 'D', 7000),
    ('B', 'C', 2400),
    ('B', 'D', 5200),
    ('C', 'D', 3000)
]

G.add_weighted_edges_from(edges)

# Create figure
plt.figure(figsize=(10, 7))

# Position nodes manually for better layout
pos = {
    'A': (0, 0),
    'B': (2, 2),
    'C': (2, -2),
    'D': (5, 0)
}

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue', 
                       node_size=1500, edgecolors='black', linewidths=2)

# Draw edges with arrows
nx.draw_networkx_edges(G, pos, edge_color='black', 
                       arrows=True, arrowsize=20, 
                       arrowstyle='-|>', width=2)

# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=14, font_weight='bold')

# Draw edge labels (distances)
edge_labels = {(u, v): f'{d}' for u, v, d in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, 
                             font_size=11, label_pos=0.5,
                             bbox=dict(boxstyle='round,pad=0.2', 
                                      facecolor='white', 
                                      edgecolor='none'))

# Set plot limits
plt.xlim(-0.5, 6)
plt.ylim(-3, 3)

# Remove axes
plt.axis('off')

# Add title
plt.title('Route Network', fontsize=16, fontweight='bold', pad=20)

# Save figure
plt.tight_layout()
plt.savefig('route_graph.png', dpi=300, bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.savefig('route_graph.pdf', bbox_inches='tight', facecolor='white')

print("Graph saved as 'route_graph.png' and 'route_graph.pdf'")
```
