```python
import numpy as np
import matplotlib.pyplot as plt

# Generate U values between 0 and 1
u = np.linspace(0, 1, 500)

# Example 1
x1_1 = u**2
x2_1 = u**3

# Example 2
x1_2 = u**2
x2_2 = (1 - u)**3

# Plot for Example 1
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(u, x1_1, label='X1 = U^2')
plt.plot(u, x2_1, label='X2 = U^3')
plt.title('Example 1')
plt.xlabel('U')
plt.ylabel('Response')
plt.legend()

# Plot for Example 2
plt.subplot(1, 2, 2)
plt.plot(u, x1_2, label='X1 = U^2')
plt.plot(u, x2_2, label='X2 = (1-U)^3')
plt.title('Example 2')
plt.xlabel('U')
plt.ylabel('Response')
plt.legend()

plt.tight_layout()
plt.savefig("q1.png")
plt.show()
```

```python
import numpy as np

n = 100000
np.random.seed(0)

# Generate U ~ Uniform(0,1)
U = np.random.uniform(0, 1, n)
U2 = np.random.uniform(0, 1, n)  # For independent case

# Example 1
X1_1 = U**2
X2_1 = U**3
X2_1_indep = U2**3

# Example 2
X1_2 = U**2
X2_2 = (1 - U)**3
X2_2_indep = (1 - U2)**3

# Covariances
print("Covariances:")
print("Example 1 CRN:", np.cov(X1_1, X2_1)[0, 1])
print("Example 2 CRN:", np.cov(X1_2, X2_2)[0, 1])

# Variance of Differences
print("\nVariance of Differences:")
print("Example 1 CRN:", np.var(X1_1 - X2_1))
print("Example 1 Indep:", np.var(X1_1 - X2_1_indep))

print("Example 2 CRN:", np.var(X1_2 - X2_2))
print("Example 2 Indep:", np.var(X1_2 - X2_2_indep))
```
