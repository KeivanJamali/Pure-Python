```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_14, x_24, x_13, x_23, x_34 = sp.symbols("x_14, x_24, x_13, x_23, x_34")
t_14, t_24 = sp.symbols("t_14, t_24")
```

```python
T_14 = 20 + 0.01 * x_14
T_24 = 20 + 0.01 * x_24
T_13 = 10 + 0.001 * x_13
T_23 = 12 + 0.001 * x_23
T_34 = 6 + 0.001 * x_34

X_14 = 10000 - 10 * t_14
X_24 = 8000 - 10 * t_24
```

```python
x = sp.symbols("x")
T_14 = 20 + 0.01 * x
T_24 = 20 + 0.01 * x
T_13 = 10 + 0.001 * x
T_23 = 12 + 0.001 * x
T_34 = 6 + 0.001 * x

X_14 = 10000 - 10 * x
X_24 = 8000 - 10 * x
```

```python
T = [[T_14, T_13 + T_34], [T_24, T_23+T_34], [X_14, X_24]]
T[0][1]
```

```python
x_values = [np.linspace(1, 10000, 10000), np.linspace(1, 10000, 10000), np.linspace(1, 100, 1000)]
titles = ["Origin=1, Destination=4", "Origin=2, Destination=4", "Demand Funcion"]
file_name = ["OD14.png", "OD24.png", "D.png"]
paths = [["path 1-4", "path 1-3-4"], ["path 2-4", "path 2-3-4"], ["Demand for 1-4", "Demand for 2-4"]]
i = -1
for pair in T:
    i += 1
    j = -1
    for form in pair:
        j += 1
        lambd_form = sp.lambdify(x, form, "numpy")
        y_values = lambd_form(x_values[i])
        plt.plot(x_values[i], y_values, label=paths[i][j])
    plt.title(titles[i])
    plt.legend()
    plt.xlabel("Demand")
    plt.ylabel("Time")
    plt.grid()
    plt.savefig(file_name[i])
    plt.show()
```
