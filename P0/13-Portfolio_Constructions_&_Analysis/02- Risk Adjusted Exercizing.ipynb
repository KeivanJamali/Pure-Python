```python
import pandas as pd
import numpy as np
```

```python
returns = pd.read_csv("02- portfolios_formed_on_ME_monthly_EW.csv", header=0, index_col=0, parse_dates=True, na_values=-99.99)
returns = returns[["Lo 20", "Hi 20"]]
returns.columns = ["smallCAP", "largeCAP"]
returns = returns/100
# returns = returns[1999:2016]
print(returns.head())
returns.plot.line()
```

```python
n_month = returns.shape[0]
annual_returns = (returns+1).prod()**(12/n_month) - 1
print(annual_returns)
```

```python
annual_vol = returns.std() * np.sqrt(12)
annual_vol
```

```python
riskfree_rate = 0.03
excess_return = annual_returns - riskfree_rate
sharp_ratio = excess_return/annual_vol
print(sharp_ratio)
```
