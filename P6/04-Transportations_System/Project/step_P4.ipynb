```python
import hashlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from modular.optimization_algorithms import GradientDescent

student_id = "403209743"

seed = int(hashlib.sha256(student_id.encode()).hexdigest(), 16)

D = 10000 + int(seed) % 10000
C = 1000
T = []
B = []
number = 4
for i in range(number):
    T.append(float((5+(2*i)) + int(seed)%(5+2*i)))
    B.append(0.1 * (i+1) + (int(seed)%10)/100)
    
def travel_time(x, t, b):
    return t * (1 + b * (x/C)**4)

def loss_fn(x: np.ndarray):
    cost = 0
    cost_cons = -D
    lambda_cons = 700
    for i in range(len(T)):
        travel_time_val = travel_time(x[i], T[i], B[i])
        cost += x[i] * travel_time_val
        cost_cons += x[i]
    return cost + lambda_cons * abs(cost_cons)

initial_point = np.random.random(number) * 100
initial_point
```

```python
tries = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500]
results = []
for t in range(len(tries)):
    result = []
    def loss_fn(x: np.ndarray):
        cost = 0
        cost_cons = -D
        lambda_cons = tries[t]
        for i in range(len(T)):
            travel_time_val = travel_time(x[i], T[i], B[i])
            cost += x[i] * travel_time_val
            cost_cons += x[i]
        return cost + lambda_cons * abs(cost_cons)
    optim = GradientDescent(loss_fn, list(initial_point), max_iteration=10000)
    x, z = optim.step()
    for i in range(len(x)):
        result.append(float(round(travel_time(x[i], T[i], B[i]), 2)))
    results.append(result)
```

```python
for res in results:
    print(res)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-10, 10, 60)
x2 = np.linspace(-10, 10, 60)
x1, x2 = np.meshgrid(x1, x2)
points = np.dstack((x1, x2))
y = np.zeros(shape=x1.shape)
for i in range(len(points)):
    for j in range(len(points[i])):
        y[i, j] = loss_fn(points[i][j])

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x1, x2, y)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("Z")
plt.tight_layout()
plt.show()
```
