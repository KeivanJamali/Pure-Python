```python
a = 1
c = 12
m = 13
x = 1
values = []
while True:
    y = (a*x + c)%m
    print(y)
    x = y
    if y in values:
        # print("repeated: ", y)
        break
    else:
        values.append(y/13)
```

```python
x_values = values[:]
y_values = values[1:]
y_values.append(0)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_values, y_values)
plt.title
```
