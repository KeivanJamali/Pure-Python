## Data Prepocessing

```python
print(nltk.corpus.stopwords.words("english"))
```

```python
# Process the text: tokenization, only alphanumeric, lower case, no stopwords
def process_text(text, boolean_search:bool=False):
    processed_text = text.lower()
    processed_text = nltk.word_tokenize(processed_text)
    processed_text = [word for word in processed_text if word.isalnum()]
    if boolean_search:
        stopwords = set(nltk.corpus.stopwords.words("english")) - {"not", "and", "or"}
    else:
        stopwords = set(nltk.corpus.stopwords.words("english"))
    processed_text = [word for word in processed_text if word not in stopwords]

    return processed_text
```

```python
processed_doc = [" ".join(process_text(doc)) for doc in documents]
processed_doc
```

```python
# Index the documents with TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_doc)
```

```python
# build a function to search the documents with TF-IDF
def search(query, vectorizer, tfidf_matrix):
    query_vec = vectorizer.transform([query])
    results = np.dot(tfidf_matrix, query_vec.T).toarray()
    return results
```

```python
# Sort and Display the function
results = search(query, vectorizer, tfidf_matrix)
results = sorted(enumerate(results), key=lambda x: x[1][0], reverse=True)

# Iterate over the sorted results to display the outcome
for idx, score in results:
    print(f"score {score[0]*100:.2f}%, Document: {documents[idx]}")
```

### Boolean search

```python
# defind a function for the boolean search
def boolean_search(query, documnets):
    query_tokens = process_text(query, boolean_search=True)

    results = []

    for doc_id, doc in enumerate(documents):
        doc_tokens = set(process_text(doc))
        include_doc = False
        
        for i, token in enumerate(query_tokens):
            if token == "and" and i + 1 < len(query_tokens):
                include_doc = include_doc and (query_tokens[i+1] in doc_tokens)

            elif token == "or" and i + 1 < len(query_tokens):
                include_doc = include_doc or (query_tokens[i + 1] in doc_tokens)

            elif token == "not" and i + 1 < len(query_tokens):
                if query_tokens[i + 1] in doc_tokens:
                    include_doc = False
                    break

            else:
                include_doc = (token in doc_tokens) or include_doc
        if include_doc:
            results.append((doc_id, doc))

    return results
```

## Pobabilitistic retrieval

```python
from rank_bm25 import BM25Okapi
```

```python
# Tokenize each cleaned document
tokenized_doc = [process_text(doc) for doc in documents]
```

```python
# Initialize the model
model = BM25Okapi(tokenized_doc)
```

```python
# defind probabilistic function
def bm25_search(query, model):
    query_tokens = process_text(query)
    doc_scores = model.get_scores(query_tokens)
    return doc_scores
```

```python
query = "I want learn toefl."
bm25_result = bm25_search(query, model)
sorted_bm25_results = np.argsort(bm25_result)[::-1]
print(bm25_result)
for idx in sorted_bm25_results:
    print(f"Score: {bm25_result[idx]:.3f} => Documnet: {documents[idx]}")
```
