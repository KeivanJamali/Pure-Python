```python
%load_ext autoreload
%autoreload 2
import Simplextep as sx
from Plot import Plot
import sympy as sp
```

```python
# Problem 2
p = Plot(use_seaborn=True, title="8x1 + 6x2")
obj_func = sp.sympify('8*x+6*y')
constraints2 = [sp.sympify('3*x + 2*y <= 28'), sp.sympify('5*x + 2*y <= 42'), sp.sympify('x <= 8'), sp.sympify('y <= 8'), sp.sympify('y >= 0'), sp.sympify('x >= 0')]
p.plot_optimization(obj_func, constraints2, minimize_obj=False)
p.set_plot_limits(x_min=-1, x_max=9, y_min=-5, y_max=11)
p.show()
p.save("Figs/Q2-a-1.svg")
```

```python
# Problem 4
objective_function = [0, 0, 0, 0]
constraints = [[2, 1, 1, 1], [6, 1, 2, 6], [2, -1, 0, 4]]
equality = ["eq", "eq", "eq"]
parameters = [("x1", "+"), ("x2", "+"), ("x3", "+")]
mode = "max"
```

```python
# Problem 5
objective_function = [0, 0, 0, 0]
constraints = [[1, 1, 1, 1], [0, 1, 1, 0]]
equality = ["eq", "eq"]
parameters = [("x1", "+"), ("x2", "+"), ("x3", "+")]
mode = "max"
```

```python
problem = sx.Problem_Prepration(objective_function=objective_function,
                                constraints=constraints,
                                equality=equality,
                                parameters=parameters,
                                mode=mode)
print(problem)
```

```python
simplex = sx.Simplex(problem=problem, number_of_dashes=12)
simplex.fit()
```
