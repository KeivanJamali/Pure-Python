```python
import matplotlib.pyplot as plt
import numpy as np

# Given PMF values
x = np.array([1, 2, 3, 4, 5])  # X values
p_x = np.array([0.1, 0.3, 0.2, 0.3, 0.1])  # Probabilities

# PMF Plot
plt.figure(figsize=(7,5))
plt.stem(x, p_x, basefmt=" ", linefmt="b-", markerfmt="bo")
plt.xlabel("X values")
plt.ylabel("P(X = x)")
plt.title("Probability Mass Function (PMF)")
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q1_1.png")
plt.show()
```

```python
# Compute CDF
F_x = np.cumsum(p_x)  # Cumulative sum

# CDF Plot
plt.figure(figsize=(7,5))
plt.step(x, F_x, where="post", color="g", label="CDF")
plt.scatter(x, F_x, color='g')
plt.xlabel("X values")
plt.ylabel("F(X)")
plt.title("Cumulative Distribution Function (CDF)")
plt.xticks(x)
plt.yticks(F_x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.savefig("q1_2.png")
plt.show()
```

```python
import sympy as sp

x = sp.symbols("x")
eq = sp.Eq(x**3 + x**2 + x, 3)
sp.solve(eq, x)
```

```python
import numpy as np
fx = x**2 + (3/2)*x + 1/3
fx_np = sp.lambdify(x, fx, "numpy")
x_values = np.linspace(0, 1, 100)
y_values = fx_np(x_values)

plt.plot(x_values, y_values, label=r"$f(x) = x^2 + \frac{3}{2}x + \frac{1}{3}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.savefig("q2_1.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given PMF values
x, y = sp.symbols("x, y")
px = (6+4*x)/30
py = (1+y)/10

x_values = np.array([0, 1, 2])  # X values
y_values = np.array([0, 1, 2, 3])
x_lamb = sp.lambdify(x, px, "numpy")
y_lamb = sp.lambdify(y, py, "numpy")
yy_values = y_lamb(y_values)
yx_values = x_lamb(x_values)

# PMF Plot
```

```python
plt.figure(figsize=(7,5))
plt.stem(y_values, yy_values, basefmt=" ", linefmt="b-", markerfmt="bo")
plt.xlabel("Y values")
plt.ylabel("P(Y = y)")
plt.title("Probability Mass Function (PMF)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q3_2.png")
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols("x, y")
fx = 3/2 - x
fy = y - 1/2

Fx  = (3*x)/2 - x**2 / 2
Fy = y**2/2 - y/2

x_values = np.linspace(0, 1, 50)
y_values = np.linspace(1, 2, 50)

fxl = sp.lambdify(x, fx)
fyl = sp.lambdify(y, fy)
Fxl = sp.lambdify(x, Fx)
Fyl = sp.lambdify(y, Fy)

fx_values = fxl(x_values)
Fx_values = Fxl(x_values)
fy_values = fyl(y_values)
Fy_values = Fyl(y_values)

```

```python
plt.figure(figsize=(7,5))
plt.plot(x_values, fx_values)
plt.xlabel("X values")
plt.ylabel("fx(X = x)")
plt.title("Probability Density Function (PDF)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q4_1.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(y_values, fy_values)
plt.xlabel("Y values")
plt.ylabel("fy(Y = y)")
plt.title("Probability Density Function (PDF)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q4_2.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(x_values, Fx_values)
plt.xlabel("X values")
plt.ylabel("Fx(X = x)")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q4_3.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(y_values, Fy_values)
plt.xlabel("Y values")
plt.ylabel("Fy(Y = y)")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("q4_4.png")
plt.show()
```
