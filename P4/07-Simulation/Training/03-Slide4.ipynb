## Discrete

### Discrete Uniform

```python
sample_discrete_uniform = DU(i=1, j=6, size=1000, print_=True)
print("mass              = ",sample_discrete_uniform.mass_function(x=3))
print("cumulative mass   = ",sample_discrete_uniform.distribution_function(x=9))
calculation = Sample_Calculation(sample_discrete_uniform)
calculation.MLE(type_="discrete_uniform")
calculation.plot_data()
calculation.plot_distribution()
```
