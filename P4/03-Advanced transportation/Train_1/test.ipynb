```python
import sympy as sp
```

```python
p, q = sp.symbols("p, q")
```

```python
p_f = 2.5*(3+200/(60-q/120))
p_f = p_f.simplify()
```

```python
TP = (p_f*q)
```

```python
MC = sp.Derivative(TP).doit().simplify()
```

```python
E1 = sp.Eq((3.8*10**4)/q, MC)
E1
```

```python
q_opt = sp.solve(E1, q)[0]
```

```python
y1 = (10**4)*3.8/q_opt
```

```python
y2 = p_f.subs(q, q_opt)
```
