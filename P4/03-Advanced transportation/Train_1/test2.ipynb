```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1, x2 = sp.symbols("x_1, x_2")
```

```python
y = sp.sin(x1/np.pi)**2 + sp.sin((x2+np.pi)/np.pi)**2
y = sp.sqrt(20-(x1**2 + x2**2))

x1_values = np.linspace(-100, 100, 10000)
# x1_values = np.sin(np.linspace(1, 100, 1000).astype(float)/np.pi)
x2_values = np.linspace(-100, 100, 10000)
# x2_values = np.sin((np.linspace(1, 100, 1000).astype(float)+np.pi)/np.pi)
lamdify = sp.lambdify([x1, x2], y, "numpy")
y_values = lamdify(x1_values, x2_values)
y_values = np.concat([-y_values, y_values])
x2_values = np.concat([x2_values, x2_values])
x1_values = np.concat([x1_values, x1_values])
```

```python
from IPython.display import clear_output


for angle2 in range(0, 180, 5):
    clear_output(wait=True)
    figure = plt.figure()
    ax = figure.add_subplot(111, projection="3d")
    ax.scatter(x1_values, x2_values, y_values, c="b", marker="o")
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.view_init(elev=45, azim=angle2)
    plt.draw()
    plt.pause(0.5)  # Pause to display the plot for half a second
    plt.clf()  # Clear the current figure
    plt.close(figure)
```
