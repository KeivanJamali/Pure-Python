```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

x = np.linspace(-5, 5, 400)

z_pdf = norm.pdf(x)
t_pdf = t.pdf(x, df=5)

plt.figure(figsize=(8,5))
plt.plot(x, z_pdf, label='Z Distribution (Standard Normal)', linewidth=2)
plt.plot(x, t_pdf, label='T Distribution (df=5)', linewidth=2, linestyle='--')

plt.title('Z vs T Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
