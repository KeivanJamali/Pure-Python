```python
# Problem 1
objective_function = [0, 2, -3, -4]
constraints = [[8, 1, 1, 3], [10, 3, 2, -1], [6, 0, 1, 0]]
equality = ["leq", "leq", "geq"]
parameters = [("x1", "+"), ("x2", "+"), ("x3", "+")]
mode = "max"
```

```python
problem = sx.Problem_Prepration(objective_function=objective_function,
                                constraints=constraints,
                                equality=equality,
                                parameters=parameters,
                                mode=mode)
print(problem)
```

```python
simplex = sx.Simplex(problem=problem, number_of_dashes=12)
simplex.fit()
```
